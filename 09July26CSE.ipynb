```python
marks = [95,82,-1,57,79,100,88,-1,76,92]
pl = []
ps=0
for stu in marks :
    if stu == -1:
        continue
    pl.append(stu)
for stu in pl:
    if stu ==100:
        break
    print(stu)

    if (90<stu<100):
        print("A grade")
        ps += 1
    elif (70<stu<90):
        print("B grade")
        ps += 1
    elif (50<stu<70):
        print("C grade")
        ps += 1
    else:
        print("D grade")
print("no of student passes",ps)

avg=sum(pl)/len(pl)
print("average marks of students",avg)
```
